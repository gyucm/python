### 함수 f(w)의 w에 대한 미분
$$
\frac{df(w)}{dw},
\frac{d}{dw}f(w), 
f'(w)
$$

### 함수 $f(w)=w^n$의 w에 대한 미분
$$
\frac{d}{dw}f(w) = \frac{d}{dw}w^n = nw^{n-1}
$$

## 중첩된 함수의 미분 : 연쇄법칙
### 대입식 미분
$f(w) = g(w)^2$<br>
$g(w) = aw+b$<br>
$f(w) = (aw+b)^2 = a^2w^2 + 2abw + b^2$<br>
$\frac{d}{dw}f(w) = 2a^2w + 2ab$<br>

### 연쇄법칙 미분
$$
\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dw}
$$

$\frac{df}{dg}$는 f를 g로 미분한다.
ex) $
\frac{df}{dg} = \frac{d}{dg}g^2=2g
$
<br>
$\frac{dg}{dw}$는 g를 w로 미분한다.
ex) $
\frac{dg}{dw} = \frac{d}{dw}(aw+b)=a
$
<br>
<br>
$
\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dw} = 2ga = 2(aw+b)a = 2a^2w+2ab
$


### 연쇄법칙 미분...확장
$$
\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dh}\cdot\frac{dh}{dw}
$$

```python
def f(w0, w1):
    return w0**2 + 2*w0*w1 + 3 #  x^2 + 2xy + 3
def df_dw0(w0, w1):
    return 2*w0 + 2*w1         #  2x + 2y
def df_dw1(w0, w1):
    return 2*w0                #  2x
#w1 = -1
w_range=10
dw=1
w0 = np.arange(-w_range, w_range+dw, dw)
w1 = np.ones(len(w0))
w1 *= -1

planes = [-1,1]
w0 = np.linspace(-w_range,w_range, 50)

figIndex = 1
plt.figure(figsize=(7,7))
plt.subplots_adjust(wspace=0.5, hspace=0.5)
for i,p in enumerate(planes):
    w1 = np.full(w0.shape, planes[i])#w1 = -1
    plt.subplot(2,2,figIndex)#plot 묘사의 위치 지정
    
    labelName = '$w_1={}$'.format(planes[i])
    plt.plot(w0,f(w0,w1), 'black', linewidth = 3, label = labelName)
    plt.plot(w0,df_dw0(w0,w1), 'cornflowerblue', linewidth = 3, label = '$dw_0$')
    plt.title(labelName)
    plt.xlabel('$w_0$')
    plt.ylabel('$y$')
    plt.ylim(-10,10)
    plt.xlim(-2,2)
    plt.grid(True)
    figIndex+=1
    plt.legend(loc='lower right')

w1 = np.linspace(-w_range,w_range, 50)    
for i,p in enumerate(planes):
    w0 = np.full(w1.shape, planes[i])#w1 = -1
    plt.subplot(2,2,figIndex)#plot 묘사의 위치 지정
    
    labelName = '$w_0={}$'.format(planes[i])
    plt.plot(w1,f(w0,w1), 'black', linewidth = 3, label = labelName)
    plt.plot(w1,df_dw1(w0,w1), 'cornflowerblue', linewidth = 3, label = '$dw_1$')
    plt.title(labelName)
    plt.xlabel('$w_1$')
    plt.ylabel('$y$')
    plt.ylim(-10,10)
    plt.xlim(-2,2)
    plt.grid(True)
    figIndex+=1
    plt.legend(loc='lower right')
    

plt.savefig('fig4_12_all.png')
```

```python
w_range = 2
dw = 0.25
w0 = np.arange(-w_range, w_range + dw, dw)
w1 = np.arange(-w_range, w_range + dw, dw)
wn = w0.shape[0]
ww0, ww1 = np.meshgrid(w0, w1)
ff = np.zeros((len(w0),len(w1)))
dff_dw0 = np.zeros((len(w0),len(w1)))
dff_dw1 = np.zeros((len(w0),len(w1)))
for i0 in range(wn):
    for i1 in range(wn):
        ff[i1,i0] = f(w0[i0],w1[i1])
        dff_dw0[i1,i0] = df_dw0(w0[i0],w1[i1])
        dff_dw1[i1,i0] = df_dw1(w0[i0],w1[i1])
        
plt.figure(figsize=(9,4))
plt.subplots_adjust(wspace=0.3)

plt.subplot(1,2,1)
cont = plt.contour(ww0, ww1, ff, 10, color='k')
cont.clabel(fmt='%0.2f', fontsize=8)
plt.xticks(range(-w_range, w_range+1,1))
plt.yticks(range(-w_range, w_range+1,1))
plt.xlim(-w_range-0.5,w_range+0.5)
plt.ylim(-w_range-0.5,w_range+0.5)
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)

plt.subplot(1,2,2)
plt.quiver(ww0, ww1, dff_dw0, dff_dw1)
plt.xticks(range(-w_range, w_range+1,1))
plt.yticks(range(-w_range, w_range+1,1))
plt.xlim(-w_range-0.5,w_range+0.5)
plt.ylim(-w_range-0.5,w_range+0.5)
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)
plt.show()
```

### 행렬
####  $$행렬 A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$$


#### $$ 역행렬 AA^{-1}=A^{-1}A=I$$
$$역행렬이란, 곱하면 단위행렬 I가 되는 행렬$$
$$A^{-1} = \frac{1}{ad-bc}\begin{bmatrix} d & -b \\ -c & a \end{bmatrix}$$

```python
import numpy as np
#vector variables ...a,b for a+b

A=np.array([[1,2,3],[4,5,6]])
print('A = ',A)
print('type = ',type(A))
print('shape A = ', A.shape)

B=np.array([[7,8,9],[10,11,12]])
print('B = ',B)
print('type = ',type(B))
print('shape B = ', B.shape)

print('A+B =',A+B)
print('A-B =',A-B)
print('2*A =',2*A)
print('A*B =',A*B)
```
