```python
import numpy as np #모듈추가
import matplotlib.pyplot as plt #모듈추가
#%matplotlib inline #모듈추가

x = np.linspace(-4,4, 100) #데이터 생성
y1 = 2**x #데이터 생성 2의 x제곱근
y2 = 3**x #데이터 생성
y3 = 0.5**x #데이터 생성

print('x = {}'.format(x)) #데이터 로그 확인
print('y1 = {}'.format(y1)) #데이터 로그 확인
print('y2 = {}'.format(y2)) #데이터 로그 확인
print('y3 = {}'.format(y3)) #데이터 로그 확인


plt.figure(figsize=(5,5)) #그래프출력
plt.plot(x,y1, 'black', linewidth = 3, label = '$y=2^x$') #라인두께 3, 하단라벨 표시 
plt.plot(x,y2, 'cornflowerblue', linewidth = 3, label = '$y=3^x$') #그래프출력
plt.plot(x,y3, 'gray', linewidth = 3, label = '$y=0.5^x$') #그래프출력
plt.ylim(-2,6) #y축 값 범위
plt.xlim(-4,4) #x축 값 범위
plt.grid(True) #격자
plt.legend(loc='lower right') #변수와 그래프 색 표시 위치
#plt.show()
plt.savefig('graph.png') #그래프출력
```

# 3차 함수 f(x)=(x-2)x(x+2)

```python
def f(x):
    y= (x-2)*x*(x+2)
    return y

print('x = 1', f(1))
print('x = 2', f(2))
print('x = 3', f(3))

x = np.array([1,2,3]) #list-> np.ndarray
x = np.arange(-3,3.5,0.5) #np.arangge -> 범위지정
x = np.linspace (-3,3,50) #start, fimish, sample수
print('x [] = {}, type = {}'.format(x, type(x)))
print(f(x))

plt.figure(figsize=(8,8)) #그래프출력
plt.plot(x,f(x),'cornflowerblue', linewidth = 3, label = '$y=(x-2)*x*(x+2)$') #그래프출력
plt.ylim(-30,30) #y축 값 범위
plt.xlim(-4,4) #x축 값 범위
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
plt.savefig('graph.png')
```

```python
def f(x):
    y= (x-2)*x*(x+2)
    return y

def f2(x,w1):
    y= (x-w1)*x*(x+2)
    return y

def f3(x,w2):
    y= (x-w2)*x*(x+2)
    return y

x = np.linspace (-3,3,50) #start, fimish, sample수
print('x [] = {}, type = {}'.format(x, type(x)))
print(f(x))

plt.figure(figsize=(5,5)) #그래프출력
plt.plot(x,f(x),'cornflowerblue', linewidth = 3, label = '$y=(x-2)*x*(x+2)$') #그래프출력
plt.plot(x,f2(x,w1=0),'black', linewidth = 3, label = '$y=(x-0)*x*(x+2)$')
plt.plot(x,f3(x,w2=4),'gray', linewidth = 3, label = '$y=(x-4))*x*(x+2)$')

plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.ylim(-30,30) #y축 값 범위
plt.xlim(-4,4) #x축 값 범위
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
plt.savefig('graph.png')
```

```python
def f3(x0,x1):
    r = 2*x0**2 + x1**2
    ans = r*np.exp(-r)
    return ans

#x0,x1에 따라서 각각 f3함수를 거쳐서 나오는 값 계산
xn= 50
x0= np.linspace(-2,2,xn) #-2~2까지 xn샘플링
x1= np.linspace(-2,2,xn) 
y= np.zeros((len(x0),len(x1)))
print('x0 = {}'.format(x0))
print('x1 = {}'.format(x1))
print('y = {}'.format(y))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0], x1[i1])

y = np.round(y,1)        
print('y = {}'.format(y))
plt.figure(figsize=(3.5,3))
plt.gray()
#plt.jet()
#plt.pink()
#plt.bone()
#plt.pcolor(y) #색
plt.contour(y)


plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
def GuGu(dan): #dan = int
    x=np.linspace(1,9,9)
    y = dan *x
    return y

x=np.linspace(1,9,9)
plt.figure(figsize=(5,5))
plt.subplots_adjust(wspace=0.5,hspace=0.5) #그래프간격

sub = 1
for i in range(3,10,2): #2행 3열
    plt.subplot(2,2,sub) #plot 묘사의 위치 지정
    plt.title('{}Dan'.format(i))
    plt.plot(x, GuGu(i),'k')
    plt.ylim(0,81)
    plt.grid(True)
    sub += 1
plt.show()  

plt.figure(figsize=(5,5))
plt.plot(x, GuGu(3),'cornflowerblue', linewidth = 3, label = '$y=3*x$')
plt.plot(x, GuGu(5),'black', linewidth = 3, label = '$y=3*x$')
plt.plot(x, GuGu(7),'gold', linewidth = 3, label = '$y=3*x$')
plt.plot(x, GuGu(9),'salmon', linewidth = 3, label = '$y=3*x$')
plt.ylim(0,91)
plt.xlim(3,9)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
